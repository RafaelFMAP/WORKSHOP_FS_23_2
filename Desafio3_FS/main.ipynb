```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dados.csv')
df.head()
```

```python
math_score = df['math score']
reading_score = df['reading score']
writing_score = df['writing score']
```

```python
math_score_analysis = {
    "min": math_score.min(),
    "max": math_score.max(),
    "mean": math_score.mean()
}

df_math = pd.DataFrame(math_score_analysis, index=[0])

plt.figure(figsize=(10, 5))
plt.bar([.5,1,1.5], df_math.values.tolist()[0], width=0.1, color=['r', 'g', 'b'])
plt.xticks([.5,1,1.5], df_math.columns.tolist())
plt.show()
df_math.head()
```

Os valores minimo, máximo e a média dos valores foram respectivamente, 0, 100 e 66,089.

```python
reading_score_analysis = {
    "min": reading_score.min(),
    "max": reading_score.max(),
    "mean": reading_score.mean()
}

df_reading = pd.DataFrame(reading_score_analysis, index=[0])

plt.figure(figsize=(10, 5))
plt.bar([.5,1,1.5], df_reading.values.tolist()[0], width=0.1, color=['r', 'g', 'b'])
plt.xticks([.5,1,1.5], df_reading.columns.tolist())
plt.show()
df_reading.head()
```

```python
writing_score_analysis = {
    "min": writing_score.min(),
    "max": writing_score.max(),
    "mean": writing_score.mean()
}

df_writing = pd.DataFrame(writing_score_analysis, index=[0])

plt.figure(figsize=(10, 5))
plt.bar([.5,1,1.5], df_writing.values.tolist()[0], width=0.1, color=['r', 'g', 'b'])
plt.xticks([.5,1,1.5], df_writing.columns.tolist())
plt.show()
df_writing.head()
```

```python
df_group_math = df[["race/ethnicity", "math score"]]
df_group_math = df_group_math.groupby("race/ethnicity", group_keys=True).mean()

plt.figure(figsize=(10, 5))
plt.plot(df_group_math.values, color='r', marker='o', linestyle='dashed', linewidth=2, markersize=6)
plt.xticks(range(len(df_group_math.index.tolist())), df_group_math.index.tolist())
plt.show()
```

```python
df_group_writing = df[["parental level of education", "writing score"]]
df_group_writing = df_group_writing.groupby("parental level of education", group_keys=True).mean()

plt.figure(figsize=(10, 5))
plt.plot(df_group_writing.values, color='r', marker='o', linestyle='dashed', linewidth=2, markersize=6)
plt.xticks(range(len(df_group_writing.index.tolist())), df_group_writing.index.tolist())
plt.show()
```

```python
df_group_reading = df[["test preparation course", "reading score"]]
df_group_reading = df_group_reading.groupby("test preparation course", group_keys=True).mean()

plt.figure(figsize=(10, 5))
plt.plot(df_group_reading.values, color='r', marker='o', linestyle='none', linewidth=2, markersize=12)
plt.xticks(range(len(df_group_reading.index.tolist())), df_group_reading.index.tolist())
plt.show()
```
